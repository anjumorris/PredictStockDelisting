### 1. READ  DATA

```python
df = pd.read_csv('../data/financial_data/financial_statements_dat.txt', delimiter ="|")
df.head()
```

```python
# load the cleaned and restructured company identity list
identity = pd.read_csv('../data/restructure_identity.csv')
identity = identity. drop(columns = 'Unnamed: 0') 
```

```python
identity.shape
```

### 2. BASIC CLEANING
Keep only the financial statement information od companies that are traded.<br>
Keep only data that that is newer tha 2014

```python
# list of company codes in the company identity list
codes = list(identity['co_code'])
```

```python
#keep only companies that are traded aka in the company identity list
df = df.loc[df['sa_finance1_cocode'].isin(codes),:] 
```

```python
df.shape
```

```python
# convert to datetime
df['sa_finance1_year'] = pd.to_datetime(df['sa_finance1_year'], format="%Y%m%d")
```

```python
# remove data that is older than 2014
df = df.loc[df['sa_finance1_year'].dt.year >= 2014,:] 
```

```python
df.shape
```

### 4. REMOVE FEATURES
1. Select only some of the features. I have made a csv file called my_mapper.csv this is used to create a column mask to select features
2. Remove columns with too many null values

open map file 'my_mapper.csv' for mapping what I need<br>
i have made my own map of features in csv to choose features that are of interest in the model 

```python
map_list = pd.read_csv('../data/financial_data/my_mapper.csv', header = None)
map_list
```

```python
# select only those columns that are in my_mapper.csv
map_list = list(map_list[0])
df = df[map_list]
```

```python
df.shape
```

```python
df.notnull().sum()
```

```python
### columns with too many nulls 
null_drop = ['sa_extra_ordi_exp','sa_extra_ordi_inc','sa_write_offs','sa_total_div',\
 'sa_equity_div_inc_dist','sa_fin_serv_exp','sa_exp_other_fin_services',\
 'sa_amortisation','sa_wip_turnover','sa_total_provisions']
```

```python
df = df.drop(columns = null_drop)
df.shape
```

```python
df.info()
```

### 5. CONVERT TO FLOAT
As one can see in the info() there are some object datatypes. This has happened as there are some ER values which we can set to nan.  

```python
# create a mask that has all object fields that need to be convered to float64 
find_obj = df.loc[:,'sa_finance1_year':'sa_sales_net_repairs_avg_nfa_net_of_reval']
msk = (find_obj.dtypes == np.object)

# mask works !!!
change_cols = find_obj.loc[:, msk].columns
    
```

```python
df['sa_pbdita_pc_total_inc'].isnull().sum()
```

```python
df['sa_pbdita_pc_total_inc'].value_counts()
```

```python
#  will covert ER to nan values
df[change_cols] = df[change_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df['sa_pbdita_pc_total_inc'].isnull().sum()
```
